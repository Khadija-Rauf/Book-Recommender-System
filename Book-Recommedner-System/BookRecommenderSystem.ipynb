# Popularity Based Recommendation System

```python
ratingsWithName = ratings.merge(books, on='ISBN')
```

```python
# Books with title and rating columns
ratingNumDf = ratingsWithName.groupby('Book-Title').count()['Book-Rating'].reset_index()
ratingNumDf.rename(columns={'Book-Rating':'ratingsNum'},inplace=True)
ratingNumDf
```

```python
#Average rating
ratingAvgDf = ratingsWithName.groupby('Book-Title').mean()['Book-Rating'].reset_index()
ratingAvgDf.rename(columns={'Book-Rating':'ratingsAvg'},inplace=True)
ratingAvgDf
```

```python
# Merge both columns
popularDf = ratingNumDf.merge(ratingAvgDf,on='Book-Title')
popularDf
```

```python
# Keep records where ratingNum is greater than 200
popularDf[popularDf['ratingsNum'] >= 200].sort_values('ratingsAvg',ascending=False)
```

```python
#To get top 100 books
popularDf = popularDf[popularDf['ratingsNum'] >= 250].sort_values('ratingsAvg',ascending=False).head(100)
popularDf
```

```python
# Merge dataset with books title and srop all duplicated
popularDf = popularDf.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','ratingsNum','ratingsAvg']]
```

# Collaborative Filtering Based Recommenedr System

```python
# Filtering users who rated minimum 200 books
x = ratingsWithName.groupby('User-ID').count()['Book-Rating'] > 200
goodUsers = x[x].index
```

```python
filteredRatings = ratingsWithName[ratingsWithName['User-ID'].isin(goodUsers)]
```

```python
#Filtering books where rating is graeter than 50 
y = filteredRatings.groupby('Book-Title').count()['Book-Rating'] >= 50
famousBooks = y[y].index
```

```python
finalRatings = filteredRatings[filteredRatings['Book-Title'].isin(famousBooks)]
```

```python
# Pivot table
pt = finalRatings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Determining euclidean distance to check similarity between books
similarityScores = cosine_similarity(pt)
```

```python
similarityScores.shape
```

```python
# It will return suggestions against given book name
def recommend(bookName):
    #fetch index with bookName
    index = np.where(pt.index==bookName)[0][0]
    #sort distances and extract 5 best ones
    similarItems = sorted(list(enumerate(similarityScores[0])), key=lambda x:x[1], reverse=True)[1:5]
    data = []
    for i in similarItems:
        item = []
        tempDf = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(tempDf.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(tempDf.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(tempDf.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('2nd Chance')
```
